```python
def get_Xtrain():
    count = 0
    b = len(sensor) // 100
    X_train =[0]*b
    for a in range(b):
        X_train[count] = []
        for i in range(0,100):
            
            X_train[count].append(sensor[i+count*100])
        count = 1 + count
    return X_train
```

```python
def get_ytrain():
    y_train = []
    b = len(sensor) // 100
    for i in range(b):
        y_train.append(sensor[i+100])
    return y_train
```
