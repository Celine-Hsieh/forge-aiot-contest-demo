```python
from flask import Flask, request
from flask_restful import Resource, Api
import requests, json, glob, time
import pandas as pd
import numpy as np

# Create Flask app object
app = Flask(__name__)
api = Api(app)

# Read CO2 csv
targetPattern = r"*.csv"
filename = glob.glob(targetPattern)[0]

# Global Variables
items = []
EEALL = [{}]
PplCnt = [0]

# GET Function
def get_CO2():
    df = pd.read_csv(filename, skiprows=18, encoding='unicode_escape', usecols=[2])  
    df = np.array(df)
    df = np.squeeze(df)
    df = df.tolist()

    CO2 = df[-1]
    return CO2

def get_SD():
    url = requests.get('http://192.168.1.197/Read_Value?0')
    text = url.text
    data = json.loads(text)
    Temp = {"name": "Temp", "value": float(data["Temp"])}
    RH = {"name": "RH", "value": float(data["RH"])}
    CO2 = {"name": "CO2", "value": int(data["CO2"])}
    PM25 = {"name": "PM25", "value": float(data["PM2.5"])}
    items = [Temp, RH, CO2, PM25]

    return items

# Resource
class SD(Resource):
    # GET Temp/RH/CO2/PM25
    def get(self, name):
        items = get_SD()
        item = next(filter(lambda x: x['name'] == name, items), None)
        value = item['value']
        return value, 200 if item else 404

class SDList(Resource):
    # GET ALL
    def get(self):
        items = get_SD()
        return {'items': items}

class CO2(Resource):
    # GET
    def get(self):
        CO2 = get_CO2()
        return CO2

class EE(Resource):
    # GET PplCnt
    def get(self):
        return PplCnt[-1]

    # POST PplCnt
    def post(self):
        data = request.get_json()
        PplCnt.pop()
        PplCnt.append(data['value'])
        localtime = time.localtime()
        time_now = time.strftime("%Y-%m-%d %I:%M:%S %p", localtime)
        returndata = str(time_now) + ', People Count: ' + str(PplCnt[-1])

        return returndata, 201

class EEA(Resource):
    # GET EEALL
    def get(self):
        return EEALL

    # POST EEALL
    def post(self):
        data = request.get_json()
        EEALL.pop()
        EEALL.append(data)

        return EEALL, 201

api.add_resource(SD, '/SD/<string:name>')
api.add_resource(SDList, '/SD')
api.add_resource(CO2, '/CO2')
api.add_resource(EE, '/EE')
api.add_resource(EEA, '/EEA')

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True)
```
